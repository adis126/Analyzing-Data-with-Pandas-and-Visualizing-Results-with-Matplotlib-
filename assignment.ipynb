```python
# ==============================================================
# Assignment: Data Loading, Analysis, and Visualization
# Objective: Use pandas & matplotlib for data exploration
# ==============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# --------------------------------------------------------------
# Task 1: Load and Explore the Dataset
# --------------------------------------------------------------

try:
    # Load Iris dataset directly from sklearn
    iris = load_iris(as_frame=True)
    df = iris.frame  # Convert to pandas DataFrame
    df['species'] = df['target'].map(dict(enumerate(iris.target_names)))

    print("‚úÖ Dataset loaded successfully!\n")
except Exception as e:
    print(f"‚ùå Error loading dataset: {e}")

# Display first few rows
print("üîπ First 5 rows of the dataset:")
display(df.head())

# Check structure, data types, and missing values
print("\nüîπ Dataset Info:")
print(df.info())

print("\nüîπ Missing values per column:")
print(df.isnull().sum())

# Clean dataset (Iris has no missing values, but we‚Äôll handle anyway)
df = df.dropna()

```

```python
# --------------------------------------------------------------
# Task 2: Basic Data Analysis
# --------------------------------------------------------------

print("\nüîπ Basic Statistics:")
display(df.describe())

# Grouping: average petal length by species
print("\nüîπ Average Petal Length by Species:")
grouped = df.groupby("species")["petal length (cm)"].mean()
print(grouped)

# Interesting finding
print("\nüìå Observation: Setosa species has noticeably smaller petals compared to Versicolor and Virginica.")
```

```python
# --------------------------------------------------------------
# Task 3: Data Visualization
# --------------------------------------------------------------

plt.style.use("seaborn-v0_8")

# 1. Line Chart (trend of sepal length across rows as an example of sequence)
plt.figure(figsize=(8,5))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length")
plt.title("Trend of Sepal Length in Dataset")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()

# 2. Bar Chart (average petal length per species)
plt.figure(figsize=(7,5))
grouped.plot(kind="bar", color=["#ff9999","#66b3ff","#99ff99"])
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Avg Petal Length (cm)")
plt.show()

# 3. Histogram (distribution of sepal width)
plt.figure(figsize=(7,5))
plt.hist(df["sepal width (cm)"], bins=15, color="#66b3ff", edgecolor="black")
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# 4. Scatter Plot (sepal length vs petal length)
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="sepal length (cm)", y="petal length (cm)", hue="species", palette="Set1")
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()

print("\n‚úÖ Analysis complete! Plots displayed successfully.")
```
